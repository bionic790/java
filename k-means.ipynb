```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
from sklearn.cluster import KMeans
```

```python
data = pd.read_csv("../input/sample12/sample1.csv")
data
```

```python
plt.scatter(data['x'],data['y'])
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
x=data.copy()
kmean=KMeans(2)
kmean.fit(x)

clusters=x.copy()
clusters['cluster_pred']=kmean.fit_predict(x)
```

```python

plt.scatter(clusters['x'],clusters['y'],c=clusters['cluster_pred'],cmap='rainbow')
plt.xlabel('x')
plt.ylabel('y')
plt.ylabel('y')
plt.show()
```
